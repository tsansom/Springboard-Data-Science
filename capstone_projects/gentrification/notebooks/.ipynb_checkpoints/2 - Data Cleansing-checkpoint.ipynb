## Missing Data
There are 10 census tracts with missing data (zeros). I'll need to figure out what to do with each of them. There are three census tracts (6 data points total) that have zeros in the rent column. This case arises when there are no properties within the census tract that are being rented at the time of the census/survey. These are pretty straight forward to impute. I'll be using a straight linear interpolation since there's so few missing data points.

### Imputable
*48453001916*: Barton Creek Habitat Preserve to Bee Caves Rd
* No rental properties - 2010, 2011, 2012

*48453001917*: Lost Creek
* No rental properties - 2014

*48491020109*: North of Lake Georgetown
* No rental properties - 2010, 2011

The remaining seven census tracts have missing data in the home value column, which is more troublesome. After digging a bit deeper with GIS and Google Maps, I've found that these tracts have no residential households, so it's no suprise that the home value data is missing. What's surprising is that they actually have some data that's not missing in the home value column. How could this be? [elaborate]

### Tracts to Drop
*48453980000*: Austin Bergstrom Airport
* No population or housing

*48453001606*: Austin State Supported Living Center
* No housing

*48453002319*: Travis County Correctional Complex
* No housing

*48453000203*: Austin State Hospital, Austin School for the Blind, Central Texas Rehabilitation Center, Texas Department of State Health Services
* No housing (some apartments)

*48453001753*: Gateway Shopping Center
* No housing (some apartments)

*48453001849*: J.J. Pickle (University of Texas), The Domain Shopping Center, Industrial Complex
* No housing (some apartments)

*48453000601*: University of Texas, Capital Complex
* No **real** housing (student apartments/dormatories and some faculty housing with hyperinflated value)

```python
#need to drop tracts before imputing because zeros will be filled with nans and interpolated otherwise
to_drop = [48453980000, 48453001606, 48453002319, 48453000203, 48453001753, 48453001849, 48453000601]
data.drop(labels=to_drop, axis=0, inplace=True)
```

```python
to_impute = [48453001916, 48453001917, 48491020109]
#replace zeros with nans so pandas knows to interpolate them
data.replace(0, pd.np.nan, inplace=True)
for geoid in to_impute:
    #taking advantage of having duplicate index entries here, usually can only use df.at[...]
    #on a single row. df.loc[...] won't work in this situation because of duplicate index entries
    data.at[geoid, 'rent_index'] = data.loc[geoid, 'rent_index'].interpolate(how='linear').values
```

```python
#now save the cleaned data
pickle_clean = open(os.path.join(os.getcwd(), '..', 'data', 'processed', 'clean_data.pickle'), 'wb')
pickle.dump(data, pickle_clean)
```
