```python
import pandas as pd
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, AffinityPropagation
from scipy.cluster.hierarchy import dendrogram, linkage

import scikitplot as skp

%matplotlib inline
```

```python
# read the pickled data
pickle_data = open(os.path.join(os.getcwd(), '..', 'data', 'processed', 'clean_data.pickle'), 'rb')
data = pickle.load(pickle_data)
# rename the columns for plotting
data.columns = ['Year', 'Income Index', 'Education Index', 'Percent White',
                'Rent Index', 'Value Index', 'Percent Employed']
```

```python
# first column is year, which I don't want to mess with
cols = data.columns[1:]
```

```python
# create instance of StandardScaler class
ss = StandardScaler()
# fit and transform the data and put back into a dataframe
data_ss = pd.DataFrame(ss.fit_transform(data[cols]), index=data.index, columns=cols)
```

```python
# create instance of PCA class
pca = PCA()
# fit and transform the data and put back into dataframe
data_pca = pd.DataFrame(pca.fit_transform(data_ss), index=data.index)
# rename the columns
data_pca.columns = ['pca{}'.format(i) for i in range(1, 7)]
```

# Clustering

```python
sil = {}
```

## K-Means

```python
def plot_kmeans(km):
    '''
    plot the kmeans clustering results on the first two principal components

    km is the kmeans model object
    '''
    # set up colors first
    cmap = plt.cm.tab10_r
    colors = cmap.colors
    # get the cluster labels and cluster centers
    clusters = km.labels_
    centers = pca.transform(km.cluster_centers_)
    # create the figure and axes
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    # plot centers first (so legend will look better)
    for cn in np.unique(km.labels_):
        axes[0].scatter(centers[cn, 0], centers[cn, 1], c=colors[cn], 
                        edgecolor='k', zorder=5, s=100, label=cn)
    # show legend with cluster center markers
    axes[0].legend()
    # now plot each point and color by cluster
    for cn in np.unique(km.labels_):
        axes[0].scatter(data_pca[data_pca['clusters'] == cn]['pca1'],
                        data_pca[data_pca['clusters'] == cn]['pca2'],
                        c=colors[cn], alpha=0.5)
    # set labels and titles
    axes[0].set(xticklabels=[], yticklabels=[])
    axes[0].set(xlabel='Principal Component 1', 
                ylabel='Principal Component 2',
                title='KMeans ({} clusters) Results'.format(km.n_clusters))
    
    # now plot the silhouette analysis on the right axis
    skp.metrics.plot_silhouette(data_ss, km.labels_, ax=axes[1], cmap=cmap)
```

```python
# create instance of KMEans class
kmeans = KMeans()
# silhouette score and within clusters sum of squares error
sil['kmeans'] = []
wcss = []
# loop 2-10 clusters to make elbow plot and silhouette score plot
for i in range(3, 11):
    # set number of clusters
    kmeans.n_clusters = i
    # fit the model
    kmeans.fit(data_ss)
    # append the wcss and sil scores
    wcss.append(kmeans.inertia_)
    sil['kmeans'].append(skp.metrics.silhouette_score(data_ss, kmeans.labels_))
    # set the clusters column in pca to cluster labels (for plotting)
    data_pca['clusters'] = kmeans.labels_
    # use the plot_kmeans function to plot clusters and silhouette analysis
    plot_kmeans(kmeans)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

_ = axes[0].plot(range(3, 11), wcss)
_ = axes[0].set(xlabel='Numer of Clusters', 
                ylabel='Within Cluster Sum of Squares',
                title="KMeans - Elbow Plot")

_ = axes[1].plot(range(3, 11), sil['kmeans'])
_ = axes[1].set(xlabel='Number of Clusters',
                ylabel='Silhouette Score',
                title='KMeans - Silhouette Scores')
```

## Hierarchical Agglomerative Clustering

```python
def plot_hac(hac):
    # set up colors first
    cmap = plt.cm.tab10_r
    colors = cmap.colors
    # get the cluster labels
    clusters = hac.labels_
    # create the figure and axes
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    # now plot each point and color by cluster
    for cn in np.unique(hac.labels_):
        axes[0].scatter(data_pca[data_pca['clusters'] == cn]['pca1'],
                        data_pca[data_pca['clusters'] == cn]['pca2'],
                        c=colors[cn], alpha=0.5)
    # set labels and titles
    axes[0].set(xticklabels=[], yticklabels=[])
    axes[0].set(xlabel='Principal Component 1', 
                ylabel='Principal Component 2',
                title='Hierarchical Agglomerative Clustering '
                      '({} clusters) Results'.format(hac.n_clusters))
    
    # now plot the silhouette analysis on the right axis
    skp.metrics.plot_silhouette(data_ss, hac.labels_, ax=axes[1], cmap=cmap)
```

```python
link = 'ward'

hac = AgglomerativeClustering(linkage=link)

sil['hac'] = []

for i in range(3, 11):
    hac.n_clusters = i
    hac.fit(data_ss)
    sil['hac'].append(skp.metrics.silhouette_score(data_ss, hac.labels_))
    data_pca['clusters'] = hac.labels_
    plot_hac(hac)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

dend = dendrogram(linkage(data_ss, method=link), ax=axes[0])
_ = axes[0].set(xticklabels=[])
_ = axes[0].set(ylabel='Distance (Euclidean)', 
                title='Dendrogram')

_ = axes[1].plot(range(3, 11), sil['hac'])
_ = axes[1].set(xlabel='Number of Clusters', ylabel='Silhouette Score',
                title='Silhouette Scores')

_ = fig.suptitle('Hierarchical Agglomerative Clustering - Ward Linkage')
```

## Affinity Propagation

```python
def plot_affin(affin):
    cmap = plt.cm.tab10_r
    colors = cmap.colors
    clusters = affin.labels_
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    for cn in np.unique(affin.labels_):
        axes[0].scatter(data_pca[clusters == cn]['pca1'],
                        data_pca[clusters == cn]['pca2'],
                        c=colors[cn], alpha=0.5)
    axes[0].set(xticklabels=[], yticklabels=[])
    axes[0].set(xlabel='Principal Component 1', 
                ylabel='Principal Component 2',
                title='Affinity Propagation Clustering '
                      '({} clusters) Results'.format(len(np.unique(clusters))))
    
    skp.metrics.plot_silhouette(data_ss, affin.labels_, ax=axes[1], cmap=cmap)
```

```python
affin = AffinityPropagation(damping=0.9)

sil['affin'] = []
nc = []
prefs = [-1000, -900, -800, -700, -600, -500, -400]

for pref in prefs:
    affin.preference = pref
    affin.fit(data_ss)
    sil['affin'].append(skp.metrics.silhouette_score(data_ss, affin.labels_))
    nc.append(len(np.unique(affin.labels_)))
    plot_affin(affin)
```

```python
plt.plot(prefs, sil['affin'])
```

```python
from sklearn.cluster import SpectralClustering

sc = SpectralClustering(n_clusters=5, eigen_solver='arpack', eigen_tol=0.00,
        affinity="nearest_neighbors")
sc.fit(data_ss)
skp.metrics.silhouette_score(data_ss, sc.labels_)
```

```python
plt.scatter(data_pca['pca1'], data_pca['pca2'], c=pd.Series(sc.labels_), alpha=0.5)
```
