````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pickle

%matplotlib inline
````

````python
pickle_data = open(os.path.join(os.getcwd(), 'data', 'processed', 'clean_data.pickle'), 'rb')
data = pickle.load(pickle_data)
data.describe()
````

````python
cols = data.columns[1:]
names = ['Income Index', 'Education Index', 'Percent White', 
         'Rent Index', 'Value Index', 'Percent Employed']
````

````python
sns.pairplot(data[cols], diag_kind='kde', diag_kws={'shade': True}, 
             plot_kws={'alpha': 0.8}, markers='.');
````

````python
pd.plotting.scatter_matrix(data[cols], figsize=(12, 12), alpha=0.3, diagonal='kde');
````

````python
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

corr = data[cols].corr()
sns.heatmap(corr, vmax=0.8, vmin=0.2, annot=True, cbar=False, square=True, ax=ax)
````

````python
fig, axes = plt.subplots(3, 2, figsize=(14, 12))

for var, ax in zip(cols, axes.flatten()):
    sns.boxplot(x='year', y=var, data=data, ax=ax, palette='Set3', fliersize=3)
````
